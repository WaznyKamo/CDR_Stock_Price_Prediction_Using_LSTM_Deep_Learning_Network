```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url_AMREST = 'https://www.bankier.pl/gielda/notowania/akcje/AMREST/komunikaty/'
```

```python
def get_announcments(stock):
    i = 1
    stock_data = []
    while True:
        print(i)
        full_url = 'https://www.bankier.pl/gielda/notowania/akcje/' + stock + '/komunikaty/' + str(i)
        response = requests.get(full_url)
        soup = BeautifulSoup(response.text, features='html.parser')
        articles = soup.find_all('div', attrs={"class": "article"})
        if articles:
            for article in articles:
                title = article.find('a').text
                date = article.find('time').text
                stock_data.append([date, title])
            i += 1
        else:
            return pd.DataFrame(stock_data, columns=['Date', 'Announcement'])
```

```python
amrest_announcements = get_announcments('AMREST')
```

```python
financial_announcemetns_Amrest = amrest_announcements.where(amrest_announcements['Annoncement']=='Wyniki finansowe').dropna()
```
